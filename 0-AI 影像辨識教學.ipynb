##### AI 影像辨識教學
本篇文章彙整了一系列 Python AI 影像辨識教學，只要按照教學文的順序閱讀和實作，就可以輕鬆入門 AI 影像辨識，\
做出人臉辨識、姿勢辨識、手勢辨識、數字辨識、物件追蹤...等影像辨識應用。

* 下列所有 AI 影像辨識都會使用 OpenCV，請先閱讀：[OpenCV 教學](https://steam.oxxostudio.tw/category/python/ai/opencv-index.html)

* 使用 MediaPipe 安裝與啟動
* Mediapipe 人臉偵測 ( Face Detection )
* Mediapipe 人臉特徵點偵測 ( Face Landmark Detection )
* Mediapipe 手掌特徵點偵測 ( Hand Landmark Detection )
